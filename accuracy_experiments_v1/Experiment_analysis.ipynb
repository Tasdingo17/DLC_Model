```python
import numpy as np
import pandas as pd
```

```python
res_file = 'exps_res/all_res.csv'
df_res = pd.read_csv(res_file)
df_skipped = df_res.loc[df_res['is_skipped'] == True]
df = df_res.loc[df_res['is_skipped'] == False]
```

```python
df
```

```python
df['p_loss_err'] = df.apply(lambda row: abs(row['p_loss'] - row['est_p_loss']), axis=1)
df['mu_err'] = df.apply(lambda row: abs(row['mu'] - row['est_mu']), axis=1)
df['e_b_err'] = df.apply(lambda row: abs(row['e_b'] - row['est_e_b']), axis=1)
df['e_gb_err'] = df.apply(lambda row: abs(row['e_gb'] - row['est_e_gb']), axis=1)
df['delay_err'] = df.apply(lambda row: abs(row['delay'] - row['est_delay']), axis=1)
df['jitter_err'] = df.apply(lambda row: abs(row['jitter'] - row['est_jitter']), axis=1)
df
```

```python
df_grouped = df.groupby(['p_loss','mu','e_b','e_gb','delay','jitter']).mean()
df_grouped
```
